```python
import pandas as pd
```

```python
def func(excelsheet_name,year):
    

    print( 'reading file...')
    df = pd.read_excel(f'{excelsheet_name}.xlsx', sheet_name=year)

    print('Separating categorical and numerical columns\n')
    cat_col=[]
    num_col=[]
    for col in df.columns:
        if df[col].dtype!='O':
            num_col.append(col)
        else:
            cat_col.append(col)
    
    print('cat_col:',cat_col)
    print('num_col:',num_col)
    print()
    
    print('Renaming columns')
    try:
        df.rename(columns = {'SubSector':'Sub Sector',num_col[0]:'GHG Emissions (kg)', num_col[1]:'Energy Intensity (eWh/HDD/sq. ft)', 
                              num_col[2]:'Energy Intensity (ekWh/ML)'}, inplace = True)
        
        #Replacing 'SubSector' as'Sub Sector'
        cat_col[1]='Sub Sector'
    except KeyError:
        df.rename(columns = {num_col[0]:'GHG Emissions (kg)', num_col[1]:'Energy Intensity (eWh/HDD/sq. ft)', 
                              num_col[2]:'Energy Intensity (ekWh/ML)'}, inplace = True)
    print(num_col[0],'renamed as','GHG Emissions (kg)')
    print(num_col[1],'renamed as','Energy Intensity (eWh/HDD/sq. ft)')
    print(num_col[2],'renamed as','Energy Intensity (ekWh/ML)')
    print()
    
    print('dataset shape')
    print (df.shape)
    
    print('Checking for dtypes,null_count,unique_values,percent_of_null_values\n')
    print(pd.DataFrame({'dtypes':df.dtypes,'null_count':df.isnull().sum(),'unique_values':df.nunique(),
                 'percent_of_null_values':df.isnull().mean()}))
    
    
    

    
    
    print('Printing unique values and their count from categorical columns')
    for col in cat_col:
        print(col,df[col].nunique())
        print(df[col].unique())
    print()
    print( '.......................................')
    
    print('Printing unique values and their count from "Operation type" column alone')
    print(df['Operation Type'].nunique())
    print(df['Operation Type'].unique())
    print()
    print( '.......................................')
    
    print('Converting values to lowercase in "Operation type" column')
    df['Operation Type']=df['Operation Type'].apply(lambda x:x.lower())
    print()
    
    print('Printing unique values and their count from "Operation type" column again... \n')
    print(df['Operation Type'].nunique())
    print(df['Operation Type'].unique())
    print()
    print( '.......................................')
    
    print('Correcting the values in "Operation type" column \n')
    a=list(df['Operation Type'])
    for op_t in a:
        if 'indoor' in op_t:
            b=a.index(op_t)
            a[b]='indoor recreational facilities'
        elif op_t=='centres communautaires':
            b=a.index(op_t)
            a[b]='community centres'
        elif op_t=="postes d'incendie et bureaux et installations connexes":
            b=a.index(op_t)
            a[b]='fire stations and associated offices and facilities'
        elif op_t=='bureaux administratifs et installations connexes, y compris les salles de réunion du conseil municipal':
            b=a.index(op_t)
            a[b]='administrative offices and related facilities, including municipal council chambers'
        elif op_t=='installations de loisir intérieures':
            b=a.index(op_t)
            a[b]='indoor recreational facilities'
        elif op_t=='patinoires intérieures':
            b=a.index(op_t)
            a[b]='indoor recreational facilities'
        elif op_t=="installations d'entreposage où le matériel et les véhicules sont entretenus, réparés ou entreposés":
            b=a.index(op_t)
            a[b]='storage facilities where equipment or vehicles are maintained, repaired or stored'
        elif op_t=="installations liées au traitement de l'eau":
            b=a.index(op_t)
            a[b]='facilities related to the treatment of water'
        elif op_t=='installations liées au pompage des eaux usées':
            b=a.index(op_t)
            a[b]='facilities related to the pumping of sewage'
        elif op_t=='installations liées au traitement des eaux usées':
            b=a.index(op_t)
            a[b]='facilities related to the treatment of sewage'
        elif op_t=="installations liées au traitement de l'eau":
            b=a.index(op_t)
            a[b]='facilities related to the treatment of water'
        elif op_t=='bureaux administratifs et installations connexes':
            b=a.index(op_t)
            a[b]='administrative offices and related facilities'
        elif op_t=='bureaux administratifs':
            b=a.index(op_t)
            a[b]='administrative offices'
        elif op_t=='stationnement intérieur':
            b=a.index(op_t)
            a[b]='indoor parking'
        elif op_t=='école':
            b=a.index(op_t)
            a[b]='school'
        elif op_t=='inoccupé':
            b=a.index(op_t)
            a[b]='unoccupied'
        elif op_t=='inoccupé':
            b=a.index(op_t)
            a[b]='unoccupied'
    
    #Adding corrected 'Operation Type' column to the dataset
    df['Operation Type']=a
    
    print()
    print('Printing unique values and their count from "Operation type" column again... \n')
    print(df['Operation Type'].nunique())
    print(df['Operation Type'].unique())
    
    #Adding year column to the dataset
    df['Year']=[f'{year}' for i in range(len(df))]

    print()
    print(f"saving as pub{year}.csv file")
    df.to_csv(f'pub{year}.csv',encoding='utf-8')
```
